```python
import pandas as pd
import numpy as np
from google.colab.patches import cv2_imshow
from matplotlib import pyplot as plt
import cv2
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
train = pd.read_csv('allemojis.csv')
```

```python
train.head()
```

```python
for i in range(0,2):
  df=train.loc[[i]]
  df.drop(['Target'],axis=1,inplace=True)
  # df
  # df.keys()
  df=df.to_numpy()
  df2=np.reshape(df,(28,-1)).astype('float32')
  cv2_imshow(df2)
```

```python
aug = ImageDataGenerator(
	rotation_range=30,
	zoom_range=0.15,
	width_shift_range=0.2,
	height_shift_range=0.2,
	shear_range=0.15,
	horizontal_flip=True,
	fill_mode="nearest")
```

```python
mp=[-1,'checkmark','cloud','croissant','heart','laugh','smile','sun']
for i in range(1,8):
  df3=train.loc[train['Target'] == str(i)]
  df2=df3.drop(['Target'],axis=1)
  df2=df2.to_numpy()
  df2=df2.reshape(-1,28,28,1)
  total=0
  for batch in aug.flow(df2, batch_size=10, save_to_dir="/content/augmented/"+mp[i],
	save_prefix="image", save_format="jpg"):
	  total += 1
	  if total == 200:
		  break
```
